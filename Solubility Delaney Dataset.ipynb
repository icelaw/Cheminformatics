Calculating the molecular descriptors 
    cLogP (Octanol-water partition coefficient)
    MW (Molecular weight)
    RB (Number of rotatable bonds)
    AP (Aromatic proportion = number of aromatic atoms/total number of heavy atoms)

```python
import numpy as np #importing libraries 
from rdkit.Chem import Descriptors 
```

```python
def generate(smiles, verbose=False): #calculating molecular descriptors

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_MolWt = Descriptors.MolWt(mol)
        desc_NumRotatableBonds = Descriptors.NumRotatableBonds(mol)
           
        row = np.array([desc_MolLogP,
                        desc_MolWt,
                        desc_NumRotatableBonds])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MolLogP","MolWt","NumRotatableBonds"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df = generate(sol.SMILES)
df
```

Generating number of aroamtic atoms 

```python
m = Chem.MolFromSmiles('COc1cccc2cc(C(=O)NCCCCN3CCN(c4cccc5nccnc54)CC3)oc21') 
#Computing for a single molecule
```

```python
aromatic_atoms = [m.GetAtomWithIdx(i).GetIsAromatic() for i in range(m.GetNumAtoms())]
aromatic_atoms
```

```python
def AromaticAtoms(m):
  aromatic_atoms = [m.GetAtomWithIdx(i).GetIsAromatic() for i in range(m.GetNumAtoms())]
  aa_count = []
  for i in aromatic_atoms:
    if i==True:
      aa_count.append(1)
  sum_aa_count = sum(aa_count)
  return sum_aa_count
```

```python
AromaticAtoms(m)
```

```python
#Computing for the whole data set
desc_AromaticAtoms = [AromaticAtoms(element) for element in mol_list]
desc_AromaticAtoms
```

Computing number of heavy atoms for the whole data set

```python
desc_HeavyAtomCount = [Descriptors.HeavyAtomCount(element) for element in mol_list]
desc_HeavyAtomCount
```

```python
desc_AromaticProportion = [AromaticAtoms(element)/Descriptors.HeavyAtomCount(element) for element in mol_list]
desc_AromaticProportion
```

```python
df_desc_AromaticProportion = pd.DataFrame(desc_AromaticProportion, columns=['AromaticProportion'])
df_desc_AromaticProportion
```

Combining the computed molecular descriptors into a data frame 

```python
df
```

```python
df_desc_AromaticProportion
```

```python
#Combining both data frames to produce the X matrix
X = pd.concat([df,df_desc_AromaticProportion], axis=1)
X
```

```python
#Y Matrix 
sol.head()

Y = sol.iloc[:,1]
Y
```

```python
#Machine Learning
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#Training and Test Data Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

```python
#Predict X_train 
Y_pred_train = model.predict(X_train)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_train, Y_pred_train))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_train, Y_pred_train))
```

```python
#Predict X_test 
Y_pred_test = model.predict(X_test)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred_test))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred_test))
```

Linear Regression Equation 

```python
yintercept = '%.2f' % model.intercept_
LogP = '%.2f LogP' % model.coef_[0]
MW = '%.4f MW' % model.coef_[1]
RB = '%.4f RB' % model.coef_[2]
AP = '%.2f AP' % model.coef_[3]
```
